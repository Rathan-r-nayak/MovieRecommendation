```python
import ast
def fetch_genres(a):
    li=[]
    for i in ast.literal_eval(a):
        li.append(i['name'])
    return li
        
```

```python
df['genres']=df['genres'].apply(fetch_genres)
```

```python
df.head(2)
```

```python
df['keywords'][0]
```

```python
df['keywords']=df['keywords'].apply(fetch_genres)
```

```python
df.head(2)
```

```python
df['cast'][1]
```

```python
def actor_ext(a):
    c=0
    li=[]
    for i in ast.literal_eval(a):
        if c==3:
            break
        c+=1
        li.append(i['name'])
    return li
```

```python
df['cast']=df['cast'].apply(actor_ext)
```

```python
df.head(2)
```

```python
df['crew'][1]
```

```python
def dirct(a):
    li=[]
    for i in ast.literal_eval(a):
        if i["job"]=="Director":
            li.append(i["name"])
    return li
```

```python
df['crew']=df['crew'].apply(dirct)
```

```python
df.head(2)
```

```python
def collapse(a):
    li=[]
    for i in a:
        li.append("".join(i.split(" ")))
    return li
```

```python
x="Sam Worthington"
print("".join(x.split(" ")))
```

```python
df['cast']=df['cast'].apply(collapse)
df['genres']=df['genres'].apply(collapse)
df['keywords']=df['keywords'].apply(collapse)
df['crew']=df['crew'].apply(collapse)
```

```python
df.head(2)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df['overview']=df['overview'].str.split()
```

```python
df.head(2)
```

```python
df['tags']=df['overview']+df['genres']+df['keywords']+df['cast']+df['crew']
```

```python
df.head(2)
```

```python
new_df=df.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
#convert text to vector

from sklearn.feature_extraction.text import CountVectorizer
```

<h2>stop_words{‘english’}, list, default=None</h2>
<h4>If ‘english’, a built-in stop word list for English is used.</h4>
<p>Stop words are words like “and”, “the”, “him”, which are presumed to be uninformative in representing the content of a text, and which may be removed to avoid them being construed as signal for prediction. Sometimes, however, similar words are useful for prediction, such as in classifying writing style or personality.</p>

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(new_df['tags']).toarray()
```

```python

from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
def recommend(mov):
    ind=new_df[new_df['title']==mov].index[0]
    dist=sorted(list(enumerate(similarity[ind])),reverse=True,key=lambda x:x[1])
    for i in range(6):
        print(new_df.iloc[dist[i][0]]['title'])
```
